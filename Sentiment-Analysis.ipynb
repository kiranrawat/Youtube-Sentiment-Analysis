```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
analyser = SentimentIntensityAnalyzer()
```

## Reading CSV Data

```python
yt_data = pd.read_csv('comments.csv')
```

```python
yt_data
```

```python
def calculate_sentiment_scores(sentence):
    score = analyser.polarity_scores(sentence)
    print("{:-<40} {}".format(sentence, str(score)))
```

```python
for ind in yt_data.index: 
     print(yt_data['VideoId'][ind], yt_data['Comments'][ind]) 
     calculate_sentiment_scores(yt_data['Comments'][ind])
```
